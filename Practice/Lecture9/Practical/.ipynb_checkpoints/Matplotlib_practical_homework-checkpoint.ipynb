### P6. Write a Python code to output the following graph. (or a similar one)
```
The productivity values for the 3 groups are the following: 

g1 = [80, 60, 80, 67, 34]
g2 = [45, 67, 23, 89, 10]
g3 = [30, 90, 50, 20, 90]
```

````python
import numpy as np


g1 = [80, 60, 80, 67, 34]
g2 = [45, 67, 23, 89, 10]
g3 = [30, 90, 50, 20, 90]


index = np.arange(5) 
bar_width = 0.15

plt.bar(index, g1, bar_width, color='g', label='Group 1')
plt.bar(index + bar_width, g2, bar_width, color='r', label='Group 2')
plt.bar(index + bar_width*2, g3, bar_width, color='b', label='Group 3')

plt.xlabel('Producticity')
plt.ylabel('Groups')
plt.title('productivity during the week')
plt.legend()


plt.show()
````
